# Converting Categorical variable into numeric 

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender', 'Married', 'Dependents', 'Education', 'Self Employed', 'Propert_Area', 'Loan_Status']
le= LabelEncoder()
for i in var_mod:
    df[i]= le.fit_transform(df[i])
```

```python
df.head(16)
```

```python
X= df.iloc [:,1:12]
Y= df.iloc [:,12]
```

```python
X.head(16)
```

```python
Y
```

# Splitting the data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size= 0.20, random_state=0)

#Here the data set is broken into two parts of the ratio of 80:20 , hence 80% is training and 20% is testing
```

```python
X_test
```

# Building Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit (X_train, Y_train)
y_predictions= model.predict (X_test)
```

```python
y_predictions
```

# Model Evaluation

```python
#import the metrics class
from sklearn import metrics
cnf_matrix= metrics.confusion_matrix(Y_test, y_predictions)
cnf_matrix
```

```python
print("Accuracy:", metrics.accuracy_score(Y_test, y_predictions))
```

# Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train, Y_train)
```
